```python
import os
import json
from transformers import RobertaTokenizerFast
```

```python
with open(os.path.join('data/wikipedia_documents.json'), "r", encoding="utf-8") as f:
    wiki = json.load(f)
```

```python
tokenizer = RobertaTokenizerFast.from_pretrained('klue/roberta-base', model_max_length=4096)
```

```python
pre_1 = [key for key, value in wiki.items() if len(tokenizer.encode_plus(value['text'], max_length=999999, padding=False)['input_ids']) < 4096]
pre_1[0]
```

```python
pre_2 = list(set(wiki[key]['text'] for key in pre_1))
pre_2[0]
```

```python
def isKorean70(input_s):
    k_count = 0
    e_count = 0
    for c in input_s:
        if ord('가') <= ord(c) <= ord('힣'):
            k_count+=1
        elif c.isalnum():
            e_count+=1
    if (k_count / (k_count + e_count)) > 0.7:
        return True
    else:
        return False
```

```python
pre_3 = [t for t in pre_2 if isKorean70(t)]
pre_3[0]
```
